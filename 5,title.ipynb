```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={'student':['prash','mane','bhumi','chandu'],
     'fees':[1000,2300,4000,5500],
     'schlorship':[100,250,30,400]}
data=pd.DataFrame(data)
data
```

```python
#adding title to the graph
```

```python
x=data.fees
y=data.schlorship
plt.plot(x,y,color='g')
plt.title('student')
plt.xlabel('fees')
plt.ylabel('schlorschip')
plt.show()
```
