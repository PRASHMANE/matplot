```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=np.arange(5)
b=[2,4,6,8,10]
c=[55,6,7,8,9]
fig=plt.figure()
ax=plt.subplot()
ax.plot(a,b,'k--',label='Frequency')
ax.plot(a,c,'-',label='periods')
ax.legend()
plt.title('frequency of a sigma')
plt.show()
```

```python
a=np.arange(5)
b=[2,4,6,8,10]
c=[55,6,7,8,9]
fig=plt.figure()
ax=plt.subplot()
ax.plot(a,b,'k--',label='Frequency')
ax.plot(a,c,'-',label='periods')
legend=ax.legend()
legend.get_frame().set_facecolor('red')
plt.title('frequency of a sigma')
plt.show()
```
