```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
student=np.array(['prash','mane','bhumi'])
marks=np.array([89,96,45])
```

```python
plt.pie(marks,labels=student,autopct='%1.2f%%')
plt.legend(labels='marks')
plt.xlabel('student')
plt.ylabel('marks')
plt.show()
```
